```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#đọc dataset
data = pd.read_csv("advertising.csv")

def correllation(x, y):
    if len(x) != len(y):
        raise ValueError("The lengths of x and y must be the same.")

    n = len(x)
    mean_x = sum(x) / n
    mean_y = sum(y) / n

    sum_xy = sum((xi - mean_x) * (yi - mean_y) for xi, yi in zip(x, y))
    sum_xx = sum((xi - mean_x) ** 2 for xi in x)
    sum_yy = sum((yi - mean_y) ** 2 for yi in y)

    if sum_xx == 0 or sum_yy == 0:
        raise ValueError("Division by zero in correlation calculation.")

    correlation_coefficient = sum_xy / ((sum_xx * sum_yy) ** 0.5)
    
    return correlation_coefficient
```

```python
x = data ['TV']
y = data ['Radio']
corr_xy = correllation(x , y )
print ( f" Correlation between TV and Sales : { corr_xy }")
```

```python
correlation_matrix = data.corr()

# Vẽ heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", linewidths=.5, cmap='coolwarm')
plt.title(f"Heatmap of Correlation Matrix (TV and Radio correlation: {corr_xy:.2f})")
plt.show()
```
